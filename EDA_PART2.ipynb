```python
fig=plt.figure(figsize=(20,10))
ax=fig.gca()
data.hist(ax=ax)
#data.hist()
#plt.show()
```

```python
#converting column Date into DateTime format
#data['Date']=pd.to_datetime(data['Date'])

#extracting year and month for each record
data['year'] = pd.DatetimeIndex(data['Date']).year
data['month'] = pd.DatetimeIndex(data['Date']).month
```

```python
data.head()
```

```python
pollutants = ['PM2.5','PM10', 'NO', 'NOx','NO2','CO','NH3','O3','SO2']
df = data.set_index('Date')
axes = df[pollutants].plot(marker='.',alpha = 0.5, linestyle = 'None', figsize = (16,20), subplots = True)
for ax in axes:
    ax.set_xlabel('Years')
    ax.set_ylabel('ug/m3')
```

Observation:-The dataset contains the information about the Air quality data in India (2015-2020).It contains air quality data and AQI (Air Quality Index) at daily level of various stations across multiple cities in India. 

```python
plt.title('Distribution of AQI Buckets 2015-2020')
sns.countplot(x='AQI_Bucket', data=data, hue='year')
```

Observations:count plot illustrated in the image shows the highest AQI index in year 2018 and 2019 and lowest in 2020.

```python
#Grouping the AQI by city and calculating the average AQI per city
x=pd.DataFrame(data.groupby(['City'])[['AQI']].mean().sort_values(by='AQI',ascending=False).head(10))
x=x.reset_index('City')

#plotting the average AQI per city
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10,5))
sns.barplot(data=x,x='AQI',y='City',orient='h',palette='viridis')
```

Observations:plotting the average AQI over the years for top 5 cities which have AQI greater than 500

```python
#plotting the average AQI over the years for top 5 cities which have AQI greater than 500

data['City'][data['AQI']>=500].value_counts()

cols=['Ahmedabad','Delhi','Patna','Gurugram','Lucknow']

for col in cols:
    plt.figure(figsize=(5,5))

    x=data[data['City']==col]
    sns.barplot(x='year',y='AQI',data=x)
    plt.title(col)
```

Observation:From above bar plots we see that for Ahmedabad highest avearge AQI is for the year 2018

```python
x=pd.DataFrame(data['City'][data['AQI']< 50].value_counts().head(10))
x=x.rename(columns={'City':'Count'})

plt.figure(figsize=(10,5))
sns.barplot(x='Count',y=x.index,data=x,palette='viridis_r')
```

Observations:Bar plot illustrating which of the cities have AQI less than 50 and plotting their value count.From this we can infer which of the cities have a relatively good air quality,Amaravati and Hyderabad having top in good air quality and so on..

```python
## AQI
print(f"Minimum and maximum range of AQI: {data.AQI.min()} - {data.AQI.max()}")
plt.title('AQI levels over 2015 to 2020')
sns.lineplot(x='year',y='AQI', data=data)
```

Observation:In the above figure, it gives the view of AQI from 2015 to 2020.In 2015 it was at a peak and year by year it went down and 2020,the air was clear.

```python
df = data[data['year'] == 2020]
plt.title('AQI Levels in the year 2020')
sns.lineplot(x='month', y='AQI', data=df);
```

```python
df = data[data['year'] == 2019]
plt.title('AQI Levels in the year 2020')
sns.lineplot(x='month', y='AQI', data=df);
```

Observation:-In above figures as compared AQI for year 2019 and 2020 so from figure it is clear that at the end of 2019 and start of 2020 AQI as went very low due to the pandemic and restriction during 2020 ,the AQI had went very low.
